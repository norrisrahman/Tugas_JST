Import Library

```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

```

Load Dataset

```python
# Load the diabetes dataset
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
```

Show Dataset

```python


# Create a pandas DataFrame for feature matrix X
X_df = pd.DataFrame(X, columns=diabetes.feature_names)

# Create a pandas DataFrame for target vector y
y_df = pd.DataFrame(y, columns=['target'])

X_df.reset_index(drop=True, inplace=True)
y_df.reset_index(drop=True, inplace=True)

out = pd.concat([X_df, y_df], axis=1)

out

# y_df
```

Scaling Data

```python
# Convert regression problem into binary classification
y_binary = np.where(y >= 150, 1, 0)

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Split Data

```python
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y_binary, test_size=0.2, random_state=42)

# Add a column of ones for the bias term
X_train = np.column_stack((np.ones(len(X_train)), X_train))
X_val = np.column_stack((np.ones(len(X_val)), X_val))
```

Learning Algorithm Using Newton-Raphson

```python
def second_order_learning(X, y, num_iterations):
    # Initialize parameters
    num_samples, num_features = X.shape
    theta = np.zeros(num_features)

    # Newton-Raphson algorithm
    for _ in range(num_iterations):
        # Calculate predictions
        y_pred = np.dot(X, theta)

        # Calculate first-order gradient and Hessian
        gradient = (1/num_samples) * np.dot(X.T, (y_pred - y))
        hessian = (1/num_samples) * np.dot(X.T, X)

        # Update parameters using Newton-Raphson formula
        theta -= np.linalg.inv(hessian) @ gradient

    return theta
```

Train Data

```python
# Set the number of iterations
num_iterations = 10

# Apply second-order learning on training data
theta = second_order_learning(X_train, y_train, num_iterations)

# Perform sample testing
sample_idx = 0  # Index of the sample to test
sample_features = X_val[sample_idx]
sample_label = y_val[sample_idx]
sample_prediction = np.dot(sample_features, theta) >= 0.5

print("Sample Features:", sample_features)
print("Sample Label:", sample_label)
print("Sample Prediction:", sample_prediction)
```
